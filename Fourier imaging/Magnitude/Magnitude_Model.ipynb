```python
# Collect all results
all_results = [results0, results1, results2, results3, results4, results5, 
               results6, results7, results8, results9, results10, results11,
               results12, results13, results14, results15]

days = list(range(16))

# Extract metrics from results
metrics_data = {
    'days': [],
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1_score': []
}

for day, result in zip(days, all_results):
    metrics_data['days'].append(day)
    metrics_data['accuracy'].append(result['accuracy'])
    # Use the metrics already calculated in results
    metrics_data['precision'].append(result['precision'])
    metrics_data['recall'].append(result['recall'])
    metrics_data['f1_score'].append(result['f1_score'])

df_metrics = pd.DataFrame(metrics_data)

# Create a comprehensive visualization
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

# 1. Line plot with all metrics
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(df_metrics['days'], df_metrics['accuracy'], marker='o', linewidth=2, 
         markersize=8, label='Accuracy', color='#2E86AB')
ax1.plot(df_metrics['days'], df_metrics['precision'], marker='s', linewidth=2, 
         markersize=8, label='Precision', color='#A23B72')
ax1.plot(df_metrics['days'], df_metrics['recall'], marker='^', linewidth=2, 
         markersize=8, label='Recall', color='#F18F01')
ax1.plot(df_metrics['days'], df_metrics['f1_score'], marker='d', linewidth=2, 
         markersize=8, label='F1-Score', color='#C73E1D')
ax1.set_xlabel('Days', fontsize=12, fontweight='bold')
ax1.set_ylabel('Score', fontsize=12, fontweight='bold')
ax1.set_title('Model Performance Across Different Days', fontsize=14, fontweight='bold')
ax1.legend(loc='best', fontsize=10)
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.set_xlim(-0.5, 15.5)
ax1.set_ylim(0, 1)

# 2. Bar plot comparing accuracy
ax2 = fig.add_subplot(gs[1, 0])
colors = plt.cm.viridis(np.linspace(0, 1, len(days)))
bars = ax2.bar(df_metrics['days'], df_metrics['accuracy'], color=colors, 
               edgecolor='black', linewidth=1.5)
ax2.set_xlabel('Days', fontsize=12, fontweight='bold')
ax2.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax2.set_title('Accuracy Comparison by Days', fontsize=14, fontweight='bold')
ax2.set_ylim(0, 1)
ax2.grid(True, alpha=0.3, axis='y', linestyle='--')
# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.3f}', ha='center', va='bottom', fontsize=8)

# 3. Heatmap of all metrics
ax3 = fig.add_subplot(gs[1, 1])
heatmap_data = df_metrics[['accuracy', 'precision', 'recall', 'f1_score']].T
sns.heatmap(heatmap_data, annot=True, fmt='.3f', cmap='YlOrRd', 
            xticklabels=days, yticklabels=['Accuracy', 'Precision', 'Recall', 'F1-Score'],
            cbar_kws={'label': 'Score'}, ax=ax3, linewidths=0.5)
ax3.set_xlabel('Days', fontsize=12, fontweight='bold')
ax3.set_title('Metrics Heatmap Across Days', fontsize=14, fontweight='bold')

# 4. Best/Worst performance comparison
ax4 = fig.add_subplot(gs[2, 0])
best_day = df_metrics.loc[df_metrics['accuracy'].idxmax(), 'days']
worst_day = df_metrics.loc[df_metrics['accuracy'].idxmin(), 'days']
comparison_days = [worst_day, best_day]
comparison_labels = [f'Worst\n(Day {int(worst_day)})', f'Best\n(Day {int(best_day)})']

x_pos = np.arange(len(comparison_labels))
width = 0.2

metrics_to_compare = ['accuracy', 'precision', 'recall', 'f1_score']
colors_comp = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']

for i, (metric, color) in enumerate(zip(metrics_to_compare, colors_comp)):
    values = [df_metrics.loc[df_metrics['days'] == day, metric].values[0] 
              for day in comparison_days]
    ax4.bar(x_pos + i*width, values, width, label=metric.replace('_', ' ').title(), 
            color=color, edgecolor='black', linewidth=1.5)

ax4.set_xlabel('Performance', fontsize=12, fontweight='bold')
ax4.set_ylabel('Score', fontsize=12, fontweight='bold')
ax4.set_title('Best vs Worst Day Performance', fontsize=14, fontweight='bold')
ax4.set_xticks(x_pos + width * 1.5)
ax4.set_xticklabels(comparison_labels)
ax4.legend(loc='upper left', fontsize=9)
ax4.set_ylim(0, 1)
ax4.grid(True, alpha=0.3, axis='y', linestyle='--')

# 5. Statistical summary
ax5 = fig.add_subplot(gs[2, 1])
ax5.axis('off')

summary_text = f"""
üìä STATISTICAL SUMMARY

Best Performing Day: Day {int(best_day)}
  ‚Ä¢ Accuracy: {df_metrics.loc[df_metrics['days'] == best_day, 'accuracy'].values[0]:.4f}
  ‚Ä¢ F1-Score: {df_metrics.loc[df_metrics['days'] == best_day, 'f1_score'].values[0]:.4f}

Worst Performing Day: Day {int(worst_day)}
  ‚Ä¢ Accuracy: {df_metrics.loc[df_metrics['days'] == worst_day, 'accuracy'].values[0]:.4f}
  ‚Ä¢ F1-Score: {df_metrics.loc[df_metrics['days'] == worst_day, 'f1_score'].values[0]:.4f}

Overall Statistics:
  ‚Ä¢ Mean Accuracy: {df_metrics['accuracy'].mean():.4f} ¬± {df_metrics['accuracy'].std():.4f}
  ‚Ä¢ Mean Precision: {df_metrics['precision'].mean():.4f} ¬± {df_metrics['precision'].std():.4f}
  ‚Ä¢ Mean Recall: {df_metrics['recall'].mean():.4f} ¬± {df_metrics['recall'].std():.4f}
  ‚Ä¢ Mean F1-Score: {df_metrics['f1_score'].mean():.4f} ¬± {df_metrics['f1_score'].std():.4f}

Performance Range:
  ‚Ä¢ Accuracy: [{df_metrics['accuracy'].min():.4f}, {df_metrics['accuracy'].max():.4f}]
  ‚Ä¢ Improvement: {((df_metrics['accuracy'].max() - df_metrics['accuracy'].min()) / df_metrics['accuracy'].min() * 100):.2f}%
"""

ax5.text(0.1, 0.5, summary_text, transform=ax5.transAxes, 
         fontsize=11, verticalalignment='center', family='monospace',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))

plt.suptitle('CNN Model Performance Analysis Across 16 Days', 
             fontsize=16, fontweight='bold', y=0.995)

plt.savefig('model_comparison_comprehensive.png', dpi=300, bbox_inches='tight')
plt.show()

# Print summary table
print("\n" + "="*80)
print("DETAILED METRICS TABLE")
print("="*80)
print(df_metrics.to_string(index=False))
print("="*80)
```
